```python
season_summary = (
    df.groupby('season',observed=True)['Sales_Volume']
    .agg(
        total_sales='sum',
        avg_sales='mean'
    )
    .reset_index()
)
```

```python

fig, ax1 = plt.subplots(figsize=(10,5))

sns.barplot(
    data=season_summary,
    x='season',
    y='total_sales',
    ax=ax1,
    color='skyblue'
)

ax1.set_xlabel("Season")
ax1.set_ylabel("Total Sales Volume", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)


ax2 = ax1.twinx()

ax2.plot(
    season_summary['season'],
    season_summary['avg_sales'],
    color='red',
    marker='o',
    linewidth=2
)

ax2.set_ylabel("Average Sales Volume", color='red')
ax2.tick_params(axis='y', labelcolor='red')

for i, row in season_summary.iterrows():
    ax2.text(
        i,
        row['avg_sales'],
        f"{row['avg_sales']:.1f}",
        color='red',
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.title("Total vs Average Sales Volume by Season")
plt.tight_layout()
plt.show()
```

```python
materials_terms_per_season = {}

for season in df['season'].unique():
    season_df = df[df['season'] == season]
    
    term_dict = {}
    
    for term in season_df['terms'].unique():
        term_materials = season_df[season_df['terms'] == term]['material'].unique().tolist()
        term_dict[term] = term_materials
    
    materials_terms_per_season[season] = term_dict

materials_terms_per_season
```

```python
summer_terms = materials_terms_per_season.get('Summer', {})
autumn_terms = materials_terms_per_season.get('Autumn', {})

comparison = {}

all_terms = set(summer_terms.keys()).union(set(autumn_terms.keys()))

for term in all_terms:
    summer_materials = set(summer_terms.get(term, []))
    autumn_materials = set(autumn_terms.get(term, []))
    
    comparison[term] = {
        'summer_only': list(summer_materials - autumn_materials),
        'autumn_only': list(autumn_materials - summer_materials),
        'common_materials': list(summer_materials & autumn_materials)
    }

comparison
```

```python
%%writefile app5.py
import streamlit as st
import plotly.express as px 

st.set_page_config(
    layout="wide",
    page_title = 'Simple DashBoard'
)

df = px.data.tips()

## side bar 
x = st.sidebar.checkbox('Show Data', False, key=1)
day = st.sidebar.selectbox("Select Day", df['day'].unique())
time = st.sidebar.selectbox('select Meal Time', df['time'].unique())


size = st.sidebar.radio('Select How many Dishes', sorted(df['size'].unique()),3, horizontal=True)


if x:
    st.header('DataSet Sample')
    st.dataframe(df.head(8))


col1, col2, col3 = st.columns([5,2, 5])
with col1:
    new_df1 = df[df['day'] == day]
    fig = px.histogram(new_df1, x = 'total_bill', color = 'sex',
                       title=f'totalt bill for {day}day'.title(), width = 700)
    st.plotly_chart(fig,use_container_width=True)
    new_df1 = df[df['size'] == size]
    fig = px.pie(new_df1, names = 'time', color = 'sex',
                 title=f'count of each meal time according to {size} dishes'.title()).update_traces(textinfo='value')
    st.plotly_chart(fig,use_container_width=True)
    
with col3:
    new_df2 = df[df['time'] == time]
    fig = px.scatter(new_df2, x='total_bill', y = 'tip', size = 'size', size_max=20,color = 'sex',
                     title=f'correlation between total bill and tips on {time}')
    st.plotly_chart(fig,use_container_width=True)
    fig = px.sunburst(df, path= ['day', 'time', 'size'], color = 'tip',
                      title=f'counting over day, time and size over tips'.title())
    st.plotly_chart(fig,use_container_width=True)
```
